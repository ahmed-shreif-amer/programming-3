```python
x = 0

def increment ():
    global x
    x += 1

def task1 (lock):
    for i in range(100):
        increment() 

def main_task ():
    global x
    x = 0 
    t1 = Thread(target=task1 , args=(lock,))
    t2 = Thread(target=task1 , args=(lock,))

    t1.start()
    t2.start()

    t1.join()
    t2.join()

if __name__ == "__main__" :
    for i in range (100): 
        main_task()
        print(f"iteration {i} = {x}")
```

### solve

```python
x = 0

def increment ():
    global x
    x += 1

def task1 (lock):
    for i in range(100):
        lock.acquire() ## make task locked shared resource 
        increment() 
        lock.release() ## make task unlock shared resource

def main_task ():
    global x
    x = 0 
    t1 = Thread(target=task1 , args=(lock,))
    t2 = Thread(target=task1 , args=(lock,))

    t1.start()
    t2.start()

    t1.join()
    t2.join()

if __name__ == "__main__" :
    for i in range (100): 
        main_task()
        print(f"iteration {i} = {x}")
```

#### events wait

```python
from threading import Event , Thread
import time

event = Event()

def task2 (event , id):
    print (f"the thread {id}, wait for signal ...")
    event.wait()
    print (f"thread {id} recived signal ...")


def main_2 ():
    t1 = Thread(target=task2 , args=(event,1))
    t2 = Thread(target=task2 , args=(event,2))
    t1.start()
    t2.start()

    print("tasks blocked")

    time.sleep(3)
    event.set()
    
    t1.join()
    t2.join()

    print("All threads finished.")

if __name__ == "__main__" :
    main_2()
```

```python
def calc_square(numbers):
    for n in numbers:
        time.sleep(5)
        print ('square '+ str(n*n))
 
def calc_cupe(numbers):
    for n in numbers:
        time.sleep(5)
        print ('cupe '+ str(n*n*n))
 
numbers = [1,3,7,22]
```

### threading

```python
t1 = Thread (target=calc_square , args=(numbers,))
t2 = Thread (target=calc_cupe , args=(numbers,))

t1.start()
t2.start()

t1.join()
t2.join()
print("done")
```

### pools

```python
import concurrent.futures

pool = concurrent.futures.ThreadPoolExecutor(max_workers=2)
pool.submit(calc_square , numbers)
pool.submit(calc_cupe , numbers)
pool.shutdown(wait=True)
```
